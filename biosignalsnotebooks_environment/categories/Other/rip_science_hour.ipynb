Respiration is an essential part of <strong class="color4">life</strong> and, thus, its assessment is fundamental when studying the human functioning in diverse settings. For example, respiration may indicate if one person is <strong class="color1">resting</strong>, <strong class='color7'>exercising</strong> and give us many other indications depending on the scope of our study.

To properly measure respiration in a quantitative point of view, there are numerous methods and sensors that can be employed, such as the <strong class="color11">Respiration Inductive Plethysmography (RIP)</strong> sensor. This sensor has a very specific way of functioning and some peculiarities that must be understood when employing it.

All the information related to the <a href="https://plux.info/sensors/317-respiration-rip-820202501.html"><strong class="color11">RIP sensor<img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></strong></a> provided by <strong class="color2">PLUX</strong> is explored in the video we present in this notebook.
